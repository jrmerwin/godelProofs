# Axiom Notation Guide for Student Retention Analysis

This guide explains the logical notation used in our axioms, making them ready for conversion to Gödel numbering.

## Basic Logical Operators

| Symbol | Meaning | Example | Natural Language |
|--------|---------|---------|------------------|
| `∧` | AND (conjunction) | `A ∧ B` | Both A and B are true |
| `∨` | OR (disjunction) | `A ∨ B` | Either A or B (or both) is true |
| `¬` | NOT (negation) | `¬A` | A is not true |
| `→` | IMPLIES (implication) | `A → B` | If A then B |
| `↔` | IF AND ONLY IF (biconditional) | `A ↔ B` | A if and only if B |
| `=` | EQUALS | `x = 1` | Variable x equals 1 |
| `>`, `<`, `≥`, `≤` | Comparisons | `x > 0.5` | x is greater than 0.5 |

## Quantifiers (for more complex axioms)

| Symbol | Meaning | Example | Natural Language |
|--------|---------|---------|------------------|
| `∀` | FOR ALL (universal) | `∀x` | For all students x |
| `∃` | EXISTS (existential) | `∃x` | There exists a student x |

## Common Patterns in Our Axioms

### 1. Simple Conditional Rules
```
Pattern: (Condition) → (Outcome)
Example: (Debtor = 1) → (Risk_high = 1)
Meaning: If a student is a debtor, then they are high risk
```

### 2. Compound Conditions
```
Pattern: (Condition1 ∧ Condition2) → (Outcome)
Example: (Debtor = 1 ∧ Tuition_fees_up_to_date = 0) → (Academic_struggle = 1)
Meaning: If a student is both a debtor AND has overdue tuition, then they will struggle academically
```

### 3. Multiple Outcome Possibilities
```
Pattern: (Condition) → (Outcome1 ∨ Outcome2)
Example: (International = 1) → (Support_needs = 1 ∨ Cultural_adjustment = 1)
Meaning: If a student is international, then they need either support or cultural adjustment (or both)
```

### 4. Threshold Rules
```
Pattern: (Variable > threshold) → (Outcome)
Example: (Curricular_units_1st_sem_enrolled - Curricular_units_1st_sem_approved > 3) → (Academic_risk = 1)
Meaning: If the gap between enrolled and approved units exceeds 3, then academic risk is high
```

### 5. Negation Rules
```
Pattern: ¬(Condition) → (Outcome)
Example: ¬(Scholarship_holder = 1) ∧ (Debtor = 1) → (Financial_stress = 1)
Meaning: If NOT a scholarship holder AND is a debtor, then financial stress is high
```

### 6. Biconditional Rules (Equivalences)
```
Pattern: (Condition) ↔ (Outcome)
Example: (Mother_qualification = high ∧ Father_qualification = high) ↔ (Family_academic_support = 1)
Meaning: High parental qualifications if and only if strong family academic support
```

## Variable Types in Our System

### Categorical Variables (Binary: 0 or 1)
- `Debtor`, `International`, `Scholarship_holder`, `Gender`, etc.
- Example: `Debtor = 1` means "is a debtor"

### Categorical Variables (Multiple values)
- `Course`, `Application_mode`, `Nationality`, etc.
- Example: `Course = "Engineering"` or coded as `Course = 1` for Engineering

### Numerical Variables
- `Age_at_enrollment`, `Curricular_units_1st_sem_enrolled`, etc.
- Used with comparison operators: `Age_at_enrollment > 25`

### Derived Variables (Created from combinations)
- `Academic_struggle_ratio = Curricular_units_1st_sem_enrolled / Curricular_units_1st_sem_approved`
- `Financial_stress = (Debtor = 1) ∧ (Tuition_fees_up_to_date = 0)`

## Examples from Student Retention Context

### Financial Risk Axiom
```
Logical Form: (Debtor = 1 ∧ Tuition_fees_up_to_date = 0 ∧ ¬(Scholarship_holder = 1)) → (Retention_risk = 1)

Natural Language: Students who are debtors, have overdue tuition, and don't have scholarships are at high retention risk
```

### Academic Preparation Axiom
```
Logical Form: (Mother_qualification ≥ higher_ed ∨ Father_qualification ≥ higher_ed) → (Academic_support = 1)

Natural Language: If either parent has higher education qualifications, the student has academic support
```

### Performance Warning Axiom
```
Logical Form: (Curricular_units_1st_sem_approved / Curricular_units_1st_sem_enrolled < 0.7) → (Academic_intervention_needed = 1)

Natural Language: If a student approves less than 70% of enrolled units, they need academic intervention
```

## Converting to Gödel Numbers

Each axiom will eventually be assigned a unique Gödel number based on:
1. **Prime factorization** of the logical structure
2. **Variable mappings** (each variable gets a unique prime number)
3. **Operator mappings** (each logical operator gets a unique prime number)

For example, a simple axiom like `(A = 1) → (B = 1)` might become:
- A = prime 2
- → = prime 3  
- B = prime 5
- Resulting in Gödel number: 2³ × 3¹ × 5² (or similar encoding)

## Tips for Reading Generated Axioms

1. **Start with the structure**: Identify the main logical pattern (implication, conjunction, etc.)
2. **Identify the variables**: Look for the actual column names from your dataset
3. **Understand the conditions**: What combination of factors triggers the outcome?
4. **Check domain logic**: Does the relationship make sense for student retention?

This notation system allows us to convert business logic into formal mathematical proofs while maintaining interpretability!

## DATA ETL

````python
import seaborn as sns

# Basic data quality checks
print("Missing values per column:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)

# Outlier detection (using simple IQR method for numeric columns)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
outlier_summary = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()
    outlier_summary[col] = outliers
print("\nOutlier count per numeric column:\n", outlier_summary)

# Simple descriptive statistics
desc_stats = df.describe()
print("\nDescriptive statistics:\n", desc_stats)

# Distributions (histograms for numeric columns)
import matplotlib.pyplot as plt
df[numeric_cols].hist(bins=20, figsize=(16, 12))
plt.tight_layout()
plt.show()

# Correlation analysis between key variables
corr = df[numeric_cols].corr()
print("\nCorrelation matrix (top correlations):\n", corr.unstack().sort_values(ascending=False)[len(numeric_cols):20])

plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

# Cross-tabulations of categorical variables with the target
categorical_cols = df.select_dtypes(include=['int64']).columns
for col in categorical_cols:
    if col != 'Target':
        ctab = pd.crosstab(df[col], df['Target'])
        print(f"\nCross-tabulation of {col} with Target:\n", ctab)

# Derived feature: academic_struggle_ratio = (enrolled units) / (approved units) for both semesters
df['academic_struggle_ratio_1st'] = df['Curricular units 1st sem (enrolled)'] / (df['Curricular units 1st sem (approved)'] + 1e-5)
df['academic_struggle_ratio_2nd'] = df['Curricular units 2nd sem (enrolled)'] / (df['Curricular units 2nd sem (approved)'] + 1e-5)
print("\nSample of new derived features:\n", df[['academic_struggle_ratio_1st', 'academic_struggle_ratio_2nd']].head())
````

````python
corr_df = corr.reset_index().melt(id_vars='index', var_name='variable', value_name='correlation')
corr_df.rename(columns={'index': 'feature_1'}, inplace=True)
corr_df.head()
````

````python
# Create lists to store the data
categories = []
dropout = []
enrolled = []
graduate = []

# For each categorical variable, get cross-tabulation with Target
for col in categorical_cols:
    ctab = pd.crosstab(df[col], df['Target'])
    for idx in ctab.index:
        categories.append(f"{col} - {idx}")
        dropout.append(ctab.loc[idx, 'Dropout'])
        enrolled.append(ctab.loc[idx, 'Enrolled'])
        graduate.append(ctab.loc[idx, 'Graduate'])

# Create the DataFrame
cross_tab_df = pd.DataFrame({
    'Category': categories,
    'Dropout': dropout,
    'Enrolled': enrolled,
    'Graduate': graduate
})

# Calculate total and percentages
cross_tab_df['Total'] = cross_tab_df['Dropout'] + cross_tab_df['Enrolled'] + cross_tab_df['Graduate']
cross_tab_df['Dropout_Rate'] = cross_tab_df['Dropout'] / cross_tab_df['Total']
cross_tab_df['Enrolled_Rate'] = cross_tab_df['Enrolled'] / cross_tab_df['Total']
cross_tab_df['Graduate_Rate'] = cross_tab_df['Graduate'] / cross_tab_df['Total']

cross_tab_df.sort_values('Total', ascending=False, inplace=True)
cross_tab_df.reset_index(drop=True, inplace=True)

cross_tab_df.head()
````

## GODEL NUMBERING AND AXIOM VALIDATION

````python
from godelNumbering import GodelNumberingSystem
from axiomValidation import AxiomValidator

# Initialize the system
godel_system = GodelNumberingSystem()
validator = AxiomValidator(godel_system)

# Load your 15 axioms from the file
with open('student_retention_axioms.txt', 'r') as f:
    axiom_text = f.read()

# Split into individual axioms (you might need to adjust this)
axioms = axiom_text.split('Axiom ')[1:]  # Skip first empty split
axioms = ['Axiom ' + axiom for axiom in axioms]

# Encode all axioms
for i, axiom in enumerate(axioms):
    godel_number = godel_system.encode_axiom(axiom, f"Axiom_{i+1}")
    print(f"Axiom {i+1}: {godel_number}")

# Validate the complete set
validation_results = validator.validate_axiom_set(axioms)
print("Validation Results:", validation_results)
````

````python
from godelHybridSearch import HybridWFFGenerator

# Create the generator with your axioms
generator = HybridWFFGenerator(godel_system, axioms)  # your 22 axioms

# THIS is what actually generates WFFs from axioms
wffs = generator.generate_wffs(depth=3, count=20, include_substitutions=True)

# Then analyze results
print(f"Generated {len(wffs)} actual WFFs:")
for i, wff in enumerate(wffs[:5], 1):
    print(f"\nWFF {i}:")
    print(f"  Formula: {wff['formula_string']}")
    print(f"  Derived from: {wff['source_axioms']}")
    print(f"  Proof steps: {len(wff['derivation_steps'])}")
    print(f"  Gödel Number: {wff['godel_number']}")
````

````python
# Check what debt-related variables actually exist
print("=== ACTUAL VARIABLES IN YOUR AXIOMS ===")
all_variables = set()
debt_related = []
graduation_related = []

for i, axiom in enumerate(generator.axiom_formulas, 1):
    print(f"Axiom {i}: {axiom.formula_string}")
    print(f"  Variables: {list(axiom.variables)}")
    
    # Collect variables
    all_variables.update(axiom.variables)
    
    # Look for debt/financial terms
    for var in axiom.variables:
        if any(term in var.lower() for term in ['debt', 'financial', 'tuition', 'scholarship']):
            debt_related.append((var, f"Axiom_{i}"))
    
    # Look for graduation/academic terms  
    for var in axiom.variables:
        if any(term in var.lower() for term in ['grad', 'academic', 'gpa', 'grade', 'success']):
            graduation_related.append((var, f"Axiom_{i}"))
    
    print()

print(f"\n=== SUMMARY ===")
print(f"Total unique variables: {len(all_variables)}")
print(f"All variables: {sorted(all_variables)}")

print(f"\nDEBT/FINANCIAL related variables:")
for var, axiom in debt_related:
    print(f"  {var} (from {axiom})")

print(f"\nGRADUATION/ACADEMIC related variables:")  
for var, axiom in graduation_related:
    print(f"  {var} (from {axiom})")
````
